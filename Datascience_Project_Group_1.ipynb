## Data Cleaning

```python

df_train['Age'] = df_train['Age'].groupby([df_train['Pclass']]).apply(lambda x: x.fillna(x.mean()))
```

```python
df_train.drop(['Cabin'],axis=1,inplace=True)
```

```python
df_train = df_train.dropna()
```

```python
df_train
```

```python
df_train.isnull().value_counts()
```

## Feature Engineering


```python
df_train['Title']=df_train['Name'].str.extract('([A-Za-z]+)\.',expand=False)
title_mapping = {"Mr": 0, "Miss": 1, "Mrs": 1, 
                 "Master": 0, "Dr": 2, "Rev": 2, "Col": 2, "Major": 2, "Mlle": 1,"Countess": 1,
                 "Ms": 1, "Lady": 1, "Jonkheer": 0, "Don": 0, "Dona" : 1, "Mme": 1,"Capt": 2,"Sir": 0 }
df_train['Title'] = df_train['Title'].map(title_mapping)
df_train.drop(['Name'],axis=1,inplace=True)
df_train.head()
```

```python


df_train.loc[df_train['Fare'] <= 17, 'Fare'] = 0
df_train.loc[(df_train['Fare'] > 17) & (df_train['Fare'] <= 30), 'Fare'] = 1
df_train.loc[(df_train['Fare'] > 30) & (df_train['Fare'] <= 100), 'Fare'] = 2
df_train.loc[ df_train['Fare'] > 100, 'Fare'] = 3
df_train.head()
```

```python
df_train.describe()
```


## Dummy variables

```python
X = pd.get_dummies(df_train['Embarked'],drop_first=True)
Y = pd.get_dummies(df_train['Sex'],drop_first=True)
```

```python
X.head()
```

```python
Y.head()
```

```python
df_train.drop(["Sex","Embarked","Ticket"],axis=1,inplace=True)
```

```python
df_train
```

```python
df_train2 = pd.concat([df_train,X,Y],axis=1)
```

```python
df_train2.head(20)
```

```python
y = df_train2['Survived']
x = df_train2.drop('Survived', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.3,random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train, y_train)
```

```python
Predicted_Value = LR.predict(x_test)
Predicted_Value
```

```python
LR.predict_proba(x_test)
```

```python
LR.coef_
```

```python
LR.intercept_
```

# Confusion Matrix and Classification Report

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, Predicted_Value)
print(cm)
accuracy_score(y_test, Predicted_Value)
```

```python
from sklearn.metrics import classification_report 
cr = classification_report(y_test, Predicted_Value)
print(cr) 
```
